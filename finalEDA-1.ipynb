```python
def get_region(district):
    if district in south:
        return 'South'
    elif district in west:
        return 'West'
    elif district in centar:
        return 'Central'
    elif district in north:
        return 'North'
    elif district in east:
        return 'East'
    else:
        return 'Unknown'
```

```python
df['المنطقة في الرياض']=df['الحي'].apply(get_region)
```
